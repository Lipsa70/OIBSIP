```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_csv("Unemployment in India.csv")
print(data.head())
```

```python
print(data.isnull().sum())
```

```python
data.columns= ["States","Date","Frequency",
               "Estimated Unemployment Rate",
               "Estimated Employed",
               "Estimated Labour Participation Rate",
               "Area"]
```

```python
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Employed", hue="States", data=data)
plt.show()
```

```python
data1=pd.read_csv("Unemployment_Rate_upto_11_2020.csv")
print(data1.head())
```

```python
data1.columns= ["States","Date","Frequency",
               "Estimated Unemployment Rate",
               "Estimated Employed",
               "Estimated Labour Participation Rate",
               "Region","longitude","latitude"]
```

```python
unemploment = data1[["States", "Region", "Estimated Unemployment Rate"]]
figure = px.sunburst(unemploment, path=["Region", "States"], 
                     values="Estimated Unemployment Rate", 
                     width=700, height=700, color_continuous_scale="RdY1Gn", 
                     title="Unemployment Rate in India")
figure.show()
```

```python
plt.figure(figsize=(10,8))
plt.title("Indian Jobless")
sns.histplot(x="Estimated Unemployment Rate", hue="Region", data=data1)
plt.show()
```
